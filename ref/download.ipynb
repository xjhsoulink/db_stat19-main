```python
import requests
import os

from pystats19.read import list_files, pull_file, load
```

```python
# Download from web.

for df_name in ['collision', 'vehicle', 'casualty']:

    url = f'https://data.dft.gov.uk/road-accidents-safety-data/dft-road-casualty-statistics-{df_name}-last-5-years.csv'
    output_dir = "./data/raw"
    output_file = os.path.join(output_dir, f"dft-road-casualty-statistics-{df_name}-last-5-years.csv")

    print(f"Downloading from {url}...")
    response = requests.get(url)
    if response.status_code == 200:
        with open(output_file, 'wb') as f:
            f.write(response.content)
        print(f"File saved to {output_file}.")
```

```python
for df_name in ['collision', 'vehicle', 'casualty']:
    raw_dir = "F:/archive/25fall/cs554/cs554_road_safety/data/raw"
    raw_file = os.path.join(raw_dir, f"dft-road-casualty-statistics-{df_name}-last-5-years.csv")
    df = load(
        raw_file,
        auto_download=False,
        convert_code_to_label=True,
        add_temporal_info=True,
        add_geo_info=True
    )

    output_dir =  "./data/cleaned"
    output_file = os.path.join(output_dir, f"dft-road-casualty-statistics-{df_name}-last-5-years.csv")
    df.to_csv(output_file)
```
